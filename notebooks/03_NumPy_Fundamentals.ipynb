## Introduction

Many statistical methods are built around vectors, matrices, and repeated numerical operations. NumPy gives Python a reliable way to represent these mathematical objects.

A NumPy array is similar to a Python list, but it is designed for numerical work. Arrays support vectorized operations, which means the same calculation can be applied to many values at once without writing an explicit loop. This is one reason NumPy is used underneath libraries such as Pandas, SciPy, Statsmodels, and Scikit-learn.

## Theory

A **NumPy array** is a fixed-type, multidimensional container for numerical data. The most common array dimensions in introductory statistics are:

- A **one-dimensional array**, which represents a vector of observations.
- A **two-dimensional array**, which represents a table or matrix.

Important array attributes include:

- `shape`: the number of elements along each dimension.
- `size`: the total number of elements.
- `dtype`: the data type stored in the array.

For statistical work, NumPy is especially useful for:

- Summarizing samples with means, medians, variances, and standard deviations.
- Centering and scaling data.
- Simulating random samples from probability distributions.
- Filtering observations with logical conditions.

## Mathematical Explanation

A one-dimensional array can represent a statistical sample:

$$\mathbf{x} = [x_1, x_2, \ldots, x_n]$$

Vectorized arithmetic applies an operation to each element. For example, centering a sample subtracts the sample mean from every observation:

$$\mathbf{x}_{centered} = \mathbf{x} - \bar{x}$$

This means:

$$[x_1 - \bar{x}, x_2 - \bar{x}, \ldots, x_n - \bar{x}]$$

Standardizing a sample expresses each value in standard deviation units:

$$z_i = \frac{x_i - \bar{x}}{s}$$

where $s$ is the sample standard deviation. NumPy lets us compute all standardized values in one expression.

## Python Examples

We will use small arrays first so each operation is easy to inspect. Then we will generate a larger simulated sample.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

### Creating Arrays

The `np.array` function converts a Python sequence into a NumPy array.

```python
scores = np.array([72, 88, 91, 64, 77, 85, 90, 73, 79, 94, 81, 68])

scores
```

```python
array_details = {
    "shape": scores.shape,
    "size": scores.size,
    "dtype": scores.dtype,
}

array_details
```

### Vectorized Operations

Vectorized operations make code shorter and closer to the mathematics. Here we add five bonus points to every score and cap the result at 100.

```python
bonus_scores = np.minimum(scores + 5, 100)

pd.DataFrame(
    {
        "original_score": scores,
        "bonus_score": bonus_scores,
    }
)
```

### Statistical Summaries

NumPy includes functions for common descriptive statistics. The `ddof=1` argument computes the sample standard deviation rather than the population standard deviation.

```python
summary = pd.DataFrame(
    {
        "statistic": ["mean", "median", "minimum", "maximum", "sample_standard_deviation"],
        "value": [
            np.mean(scores),
            np.median(scores),
            np.min(scores),
            np.max(scores),
            np.std(scores, ddof=1),
        ],
    }
)

summary
```

### Centering and Standardizing

Centering subtracts the mean. Standardizing subtracts the mean and divides by the standard deviation.

```python
centered_scores = scores - np.mean(scores)
z_scores = centered_scores / np.std(scores, ddof=1)

pd.DataFrame(
    {
        "score": scores,
        "centered_score": centered_scores,
        "z_score": z_scores.round(2),
    }
)
```

The mean of centered scores is approximately zero. Very small nonzero results can appear because computers store decimal values with finite precision.

```python
centered_scores.mean().round(10)
```

### Two-Dimensional Arrays

A two-dimensional array can represent rows and columns. In the array below, each row is a student and each column is an assessment.

```python
assessment_scores = np.array(
    [
        [78, 82, 85],
        [91, 88, 94],
        [64, 70, 73],
        [85, 89, 87],
        [72, 75, 79],
    ]
)

assessment_scores
```

```python
student_means = assessment_scores.mean(axis=1)
assessment_means = assessment_scores.mean(axis=0)

student_means, assessment_means
```

The `axis` argument tells NumPy which dimension to summarize. With `axis=0`, NumPy summarizes down rows for each column. With `axis=1`, NumPy summarizes across columns for each row.

### Boolean Masks

A Boolean mask filters an array using a condition. This is useful for selecting observations that meet a statistical or practical criterion.

```python
high_scores = scores[scores >= 85]

high_scores
```

### Random Number Generation

Simulation is central to modern statistics. A random generator with a fixed seed makes the result reproducible.

```python
rng = np.random.default_rng(seed=42)
simulated_heights = rng.normal(loc=170, scale=8, size=200)

simulated_heights[:10].round(1)
```

## Visualizations

The histogram below shows the simulated height distribution. The vertical line marks the simulated sample mean.

```python
height_mean = simulated_heights.mean()

fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(simulated_heights, bins=16, color="#4C78A8", edgecolor="white")
ax.axvline(height_mean, color="#F58518", linewidth=2, label=f"Mean = {height_mean:.1f}")
ax.set_title("Simulated Heights")
ax.set_xlabel("Height in centimeters")
ax.set_ylabel("Frequency")
ax.legend()
plt.show()
```

A second plot shows standardized scores. Values above zero are above the sample mean; values below zero are below the sample mean.

```python
fig, ax = plt.subplots(figsize=(8, 5))
positions = np.arange(scores.size)
colors = np.where(z_scores >= 0, "#54A24B", "#E45756")

ax.bar(positions, z_scores, color=colors)
ax.axhline(0, color="black", linewidth=1)
ax.set_title("Standardized Exam Scores")
ax.set_xlabel("Student index")
ax.set_ylabel("z-score")
plt.show()
```
