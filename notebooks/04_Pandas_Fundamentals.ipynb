## Introduction

Most real statistical projects begin with data in table form. Each row usually represents an observation, such as a student, patient, transaction, country, or experiment trial. Each column usually represents a variable, such as age, score, treatment group, income, or measurement time.

Pandas helps us move from raw tables to analysis-ready data. It is especially useful for checking data quality, selecting variables, filtering observations, computing summaries, and preparing inputs for visualization and modeling.

## Theory

Pandas has two core data structures:

- A **Series** is a one-dimensional labeled array. It is useful for representing a single variable.
- A **DataFrame** is a two-dimensional labeled table. It is useful for representing a dataset with rows and columns.

Statistical work with tabular data often includes:

- **Inspection**: checking the first rows, column names, data types, and missing values.
- **Selection**: choosing relevant columns or rows.
- **Transformation**: creating new variables or modifying existing variables.
- **Aggregation**: reducing many rows into summary statistics.
- **Grouping**: calculating summaries separately for categories.

A good Pandas workflow keeps the original analytical question in view. The purpose is not to manipulate tables for their own sake, but to make the data reliable and informative.

## Mathematical Explanation

A DataFrame can be viewed as a matrix-like object with rows and columns:

$$X = \begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1p} \\
x_{21} & x_{22} & \cdots & x_{2p} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \cdots & x_{np}
\end{bmatrix}$$

Here, $n$ is the number of observations and $p$ is the number of variables.

For a numerical column $x$, the sample mean is:

$$\bar{x} = \frac{1}{n}\sum_{i=1}^{n}x_i$$

For a grouped mean, observations are split into groups. If group $g$ contains $n_g$ observations, then the group mean is:

$$\bar{x}_g = \frac{1}{n_g}\sum_{i \in g}x_i$$

Pandas `groupby` operations implement this idea by splitting data into groups, applying a summary function, and combining the results.

## Python Examples

We will use a small student dataset created inside the notebook. This keeps the example reproducible and avoids external file dependencies.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

### Creating a DataFrame

A DataFrame can be created from a dictionary. Each key becomes a column name, and each list becomes the data for that column.

```python
students = pd.DataFrame(
    {
        "student_id": range(1, 13),
        "section": ["A", "A", "B", "B", "A", "B", "A", "B", "A", "B", "A", "B"],
        "study_hours": [3.5, 4.0, 2.0, 5.5, 6.0, 4.5, 3.0, 7.0, 5.0, 4.0, 6.5, 2.5],
        "attendance_rate": [0.86, 0.91, 0.74, 0.95, 0.98, 0.89, 0.80, 0.97, 0.93, 0.84, 0.96, np.nan],
        "exam_score": [72, 78, 65, 86, 91, 82, 70, 95, 88, 79, 93, 68],
    }
)

students
```

### Inspecting Data

Before calculating statistics, inspect the dataset. This helps catch missing values, unexpected data types, and basic structure problems.

```python
students.head()
```

```python
students.dtypes
```

```python
students.isna().sum()
```

The `attendance_rate` column contains one missing value. Missing data should never be ignored automatically; the correct treatment depends on the data source and analysis goal.

### Selecting Columns and Filtering Rows

Column selection and row filtering are common steps in statistical analysis.

```python
students[["study_hours", "exam_score"]].head()
```

```python
high_scoring_students = students[students["exam_score"] >= 85]

high_scoring_students
```

### Creating New Variables

New variables are often created to support interpretation. Here we create a pass indicator and a z-score version of exam score.

```python
students = students.assign(
    passed=students["exam_score"] >= 70,
    exam_z_score=(students["exam_score"] - students["exam_score"].mean()) / students["exam_score"].std(ddof=1),
)

students.head()
```

### Descriptive Statistics

The `describe` method gives a quick numerical summary of each numerical column.

```python
students[["study_hours", "attendance_rate", "exam_score"]].describe()
```

### Grouped Summaries

Grouped summaries compare statistics across categories. Here we summarize performance by class section.

```python
section_summary = students.groupby("section").agg(
    student_count=("student_id", "count"),
    mean_study_hours=("study_hours", "mean"),
    mean_exam_score=("exam_score", "mean"),
    pass_rate=("passed", "mean"),
)

section_summary
```

The pass rate is the mean of a Boolean variable. In Python, `True` behaves like 1 and `False` behaves like 0 in numerical summaries.

### Sorting Data

Sorting helps identify high and low observations.

```python
students.sort_values("exam_score", ascending=False).head(5)
```

## Visualizations

Pandas works closely with Matplotlib. The next plot compares exam scores by section.

```python
fig, ax = plt.subplots(figsize=(8, 5))
students.boxplot(column="exam_score", by="section", ax=ax, grid=False)
ax.set_title("Exam Scores by Section")
ax.set_xlabel("Section")
ax.set_ylabel("Exam score")
fig.suptitle("")
plt.show()
```

A scatter plot helps us inspect the relationship between study hours and exam score.

```python
fig, ax = plt.subplots(figsize=(8, 5))

for section, group in students.groupby("section"):
    ax.scatter(group["study_hours"], group["exam_score"], s=70, label=f"Section {section}")

ax.set_title("Study Hours and Exam Scores")
ax.set_xlabel("Study hours")
ax.set_ylabel("Exam score")
ax.legend(title="Class section")
plt.show()
```
