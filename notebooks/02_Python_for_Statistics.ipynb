## Theory

A programming language gives us a way to represent data and operations. For statistics, the most important Python concepts are:

- **Objects**: values stored in memory, such as numbers, text, lists, arrays, and tables.
- **Variables**: names that refer to objects.
- **Expressions**: combinations of values and operators that Python can evaluate.
- **Functions**: reusable blocks of code that take inputs and return outputs.
- **Control flow**: tools such as `if` statements and `for` loops that control what code runs.
- **Libraries**: external packages that extend Python. In this course, the most common libraries are NumPy, Pandas, Matplotlib, SciPy, Statsmodels, and Scikit-learn.

Good statistical programming emphasizes clarity. Code should make the analytical question visible, not hide it.

## Mathematical Explanation

Many statistical calculations can be written as operations over collections of values. Suppose we observe a sample:

$$x_1, x_2, \ldots, x_n$$

The sample mean is:

$$\bar{x} = \frac{x_1 + x_2 + \cdots + x_n}{n}$$

In programming terms, this is a sum divided by a count:

$$\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}$$

Python lets us translate this formula directly into code. A careful translation should preserve the logic of the formula while making the calculation reproducible.

## Python Examples

We will begin with built-in Python objects, then move to NumPy arrays and Pandas DataFrames.

```python
import math
import statistics

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

### Numbers, Variables, and Expressions

Variables let us name values so that calculations are easier to read.

```python
sample_size = 12
total_score = 936
mean_score = total_score / sample_size

mean_score
```

### Lists and Basic Summaries

A Python list can store a sequence of observations. Built-in functions such as `sum` and `len` are enough to compute a mean.

```python
study_hours = [3.5, 4.0, 2.0, 5.5, 6.0, 4.5, 3.0, 7.0, 5.0, 4.0, 6.5, 2.5]

manual_mean = sum(study_hours) / len(study_hours)
library_mean = statistics.mean(study_hours)

manual_mean, library_mean
```

### Functions

Functions make repeated calculations easier to reuse and test. The function below computes a sample range.

```python
def sample_range(values):
    """Return the difference between the largest and smallest values."""
    return max(values) - min(values)

sample_range(study_hours)
```

### Conditionals and Loops

Conditionals and loops help us classify or transform observations. Here we label each student as above or below the sample mean study time.

```python
labels = []

for hours in study_hours:
    if hours >= manual_mean:
        labels.append("at_or_above_mean")
    else:
        labels.append("below_mean")

labels[:6]
```

### NumPy Arrays

NumPy arrays are designed for numerical computing. They allow concise vectorized operations over many values.

```python
hours_array = np.array(study_hours)
centered_hours = hours_array - hours_array.mean()

pd.DataFrame(
    {
        "study_hours": hours_array,
        "centered_hours": centered_hours,
    }
).head()
```

### Pandas DataFrames

Pandas DataFrames represent tabular data. Each column is a variable, and each row is an observation.

```python
students = pd.DataFrame(
    {
        "student_id": range(1, 13),
        "study_hours": study_hours,
        "exam_score": [72, 78, 65, 86, 91, 82, 70, 95, 88, 79, 93, 68],
        "section": ["A", "A", "B", "B", "A", "B", "A", "B", "A", "B", "A", "B"],
    }
)

students
```

```python
students.groupby("section")["exam_score"].agg(["count", "mean", "std"])
```

## Visualizations

The scatter plot below shows the relationship between study time and exam score in this small example dataset.

```python
fig, ax = plt.subplots(figsize=(8, 5))

for section, group in students.groupby("section"):
    ax.scatter(group["study_hours"], group["exam_score"], label=f"Section {section}", s=70)

ax.set_title("Study Hours and Exam Scores")
ax.set_xlabel("Study hours")
ax.set_ylabel("Exam score")
ax.legend(title="Class section")
plt.show()
```

A simple fitted line can help us summarize the overall direction of the relationship. Later notebooks will explain regression in detail.

```python
slope, intercept = np.polyfit(students["study_hours"], students["exam_score"], deg=1)
x_values = np.linspace(students["study_hours"].min(), students["study_hours"].max(), 100)
y_values = intercept + slope * x_values

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(students["study_hours"], students["exam_score"], color="#4C78A8", s=70)
ax.plot(x_values, y_values, color="#F58518", linewidth=2)
ax.set_title("Study Hours and Exam Scores with Linear Trend")
ax.set_xlabel("Study hours")
ax.set_ylabel("Exam score")
plt.show()
```
